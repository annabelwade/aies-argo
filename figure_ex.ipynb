```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_params import mpl_params

plt.rcParams.update(mpl_params(size=16, font_family='DejaVu Serif', mathtext='cm'))
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Sine Wave', color='blue', linewidth=2)
plt.grid(True)
plt.legend()
plt.title(r'sine wave $y = \sin(x)$')
plt.show()
```
